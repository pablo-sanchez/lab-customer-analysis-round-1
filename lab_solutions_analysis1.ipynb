```python
#8.Perform the data cleaning operations mentioned so far in class
#8.1.Delete the column education and the number of open complaints from the dataframe.
```

```python
data = data.drop(['education', 'number_of_open_complanints'], axis=1)
```

```python
#8.2.Correct the values in the column customer lifetime value. 
#They are given as a percent, so multiply them by 100 and change dtype to numerical type.
```

```python
data['customer_lifetime_value'] = pd.to_numeric(data['customer_lifetime_value'], errors='coerce')
data['customer_lifetime_value'] = data['customer_lifetime_value']/100
```

```python
#8.3.Check for duplicate rows in the data and remove if any.
```

```python
data = data.drop_duplicates()
data.shape
```
